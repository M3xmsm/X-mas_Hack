```python
import pandas as pd
```

```python
## Объявляем черные списки провайдеров, ip и bin. 

bl_providers = ['23759997b3c59884dc4c0ff5320d6301b0e7f63bf0f6483a7b54d7d43bc5ccd1']
bl_ip = ['d1f89050054d43731ef39fc602eedac5a6cb72011293845dbf6cc4d67741f5fa']
bl_bin = ['aefb8cf11eedb0899f8b413f90cf20ebaaedcf5d300c95cceea3c924']
```

```python
## Объявляем функцию для проверки операций по черным спискам

def check_bl(df):
    df_p_checked = df.loc[~df['providerId'].isin(bl_providers)]
    df_clean = df_p_checked.loc[(~df_p_checked['ip'].isin(bl_ip)) | (~df_p_checked['bin_hash'].isin(bl_bin))]
    return df_clean
```
